```python
import pandas as pd
import numpy as np
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

**Data collection and peprocessing**

```python
dataset = sklearn.datasets.load_breast_cancer()
print(dataset)
```

```python
# load data into a data frame
data_frame = pd.DataFrame(dataset.data, columns=dataset.feature_names)
data_frame.head()
```

```python
# checking for all information
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# Adding the target coloumn
data_frame['label'] = dataset.target
data_frame.tail()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# check the distribution of the target variables
data_frame['label'].value_counts()
```

```python
# group values accoridng to the lables and find the average of each group
data_frame.groupby('label').mean()
```

**Seperating the features and target**

```python
X = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
print(X)
```

```python
print(y)
```

Splitting the data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=2) # creating 4 arrays and assing the splitted data
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Standardize the data ( to increase the accuracy)**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std =- scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test) # use th same scaler to standardize the testing set
```

```python
print(X_train_std)
```

**Buidling the Neural Network**

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of the NN

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),  # Input layer
    keras.layers.Dense(20, activation='relu'),  # Hidden layer
    keras.layers.Dense(2, activation='sigmoid')  # Output layer
])
```

```python
# compiling the neural network

model.compile (optimizer='adam',
               loss = 'sparse_categorical_crossentropy',
               metrics = ['accuracy']
               )
```

```python
# training the NN
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

**Visualizing accuracy and loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

**Get the accuracy of the model on test data**

```python
loss, accuracy_val = model.evaluate(X_test_std, Y_test)
print(accuracy_val)
```

**Building a predicative system based on this NN**

```python
input_data = (11.9,14.65,78.11,432.8,0.1152,0.1296,0.0371,0.03003,0.1995,0.07839,0.3962,0.6538,3.021,25.03,0.01017,0.04741,0.02789,0.0111,0.03127,0.009423,13.15,16.51,86.26,509.6,0.1424,0.2517,0.0942,0.06042,0.2727,0.1036)

# change the input data to a numpy array
input_data_As_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predictiong for only one data point
reshaped_input_data = input_data_As_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(reshaped_input_data)

prediction = model.predict(std_data)

print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if (prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
